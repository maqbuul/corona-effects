```python
import pandas as pd
import plotly.graph_objects as go
import plotly.subplots as sp
```

```python
df = pd.read_excel('/content/drive/MyDrive/DataSets/G20_Indices_and_Eco_Policy_Uncertainty.xlsx')
```

```python
death_data = pd.read_excel('/content/drive/MyDrive/DataSets/G20_Indices_and_Eco_Policy_Uncertainty.xlsx', sheet_name='Covid_deaths')
```

```python
death_data.columns
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')
```

```python
df_data = df[['Date', 'CountryName', 'StringencyIndex_WeightedAverage_ForDisplay', 'GovernmentResponseIndex_WeightedAverage_ForDisplay', 'ContainmentHealthIndex_WeightedAverage_ForDisplay', 'EconomicSupportIndex_ForDisplay']]
```

```python
covid_deaths = death_data[['continent','location','date','total_deaths','new_deaths']]
```

```python
df_data['CountryName'].unique()
```

**Simple Moving Average(14 days) of USA**

```python
us_data = df_data[df_data['CountryName'] == 'United States'].reset_index()
```

```python
us_deaths = covid_deaths[covid_deaths['location'] == 'United States'].reset_index()
```

```python
us_deaths['new_deaths_smoothed'] = us_deaths['new_deaths'].rolling(window=14).mean()
```

```python
us_data['StringencyIndex'] = us_data['StringencyIndex_WeightedAverage_ForDisplay'].rolling(window=14).mean()
us_data['GovernmentResponseIndex'] = us_data['GovernmentResponseIndex_WeightedAverage_ForDisplay'].rolling(window=14).mean()
us_data['ContainmentHealthIndex'] = us_data['ContainmentHealthIndex_WeightedAverage_ForDisplay'].rolling(window=14).mean()
us_data['EconomicSupportIndex'] = us_data['EconomicSupportIndex_ForDisplay'].rolling(window=14).mean()
```

```python
# this graph shows the deaths data of 3 years 2020-2022
fig = go.Figure()

fig.add_trace(go.Scatter(x=us_deaths['date'],
                         y=us_deaths['new_deaths_smoothed'],
                         name='New_deaths',
                         mode='lines',
                         line=dict(color='magenta')))

fig.update_layout(title='Deaths Simple Moving Average for United States\nFrom 2020 - 2022',
                  xaxis=dict(title='Date',dtick='M1',tickangle=-90),
                  yaxis=dict(title='New_deaths'),
                  showlegend=True,
                  xaxis_range=['2020-01-01', '2022-12-31'],
                  margin=dict(l=50, r=50, t=50, b=50),
                  width=1100,
                  height=500)

fig.show()
```
